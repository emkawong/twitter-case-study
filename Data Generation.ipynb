```python
data['top20hashtags'][:5]
```

```python
def init_dates(df, ty = 'hashtag'):
    c2 = {'hashtag': 'hash', 'emoji': 'emo'}[ty]
    rows = (df[['date', ty]]
             .withColumn(c2, explode(df[ty]))[['date', c2]]
            )
    return rows

def finish_dates(df, tag, ty = 'hashtag'):
    c2 = {'hashtag': 'hash', 'emoji': 'emo'}[ty]
    rows = df.filter(f"{c2} == '{tag}'")[['date']].collect()
    return [row.date for row in rows]
```

```python
rowshash = init_dates(df)
rowshash.cache()

for row in data['top20hashtags'][5:10]:
    s = row[0]
    data['timestamps'][s] = [time.strftime('%Y-%m-%d %H:%M:%S', time.strptime(t,'%a %b %d %H:%M:%S +0000 %Y')) for t in finish_dates(rowshash, s)]
```

```python
rows = init_dates(df, ty='emoji')
rows.cache()

for row in data['top20emoji'][5:10]:
    s = row[0]
    data['timestamps'][s] = [time.strftime('%Y-%m-%d %H:%M:%S', time.strptime(t,'%a %b %d %H:%M:%S +0000 %Y')) for t in finish_dates(rows, s, ty='emoji')]
```
