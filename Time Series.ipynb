```python
import pickle
data = pickle.load( open( "data/results.p", "rb" ) )
```

```python
import plotly.graph_objects as go

def plotly_compare(ty = 'hashtags', amt=5, l=None):
    fig = go.Figure()
    
    if l is None:
        l = [x[0] for x in data[f'top20{ty}'][:amt]]
    
    for title in l:
        fig.add_trace(go.Histogram(
            x=data['timestamps'][title],
            name=title, # name used in legend and hover labels
            xbins=dict( # bins used for histogram
                size=1000*60*60*1
            ),
            opacity=0.75,
        ))

    fig.update_layout(
        title_text=f'Most popular {ty}', # title of plot
        xaxis_title_text='Timestamp', # xaxis label
        yaxis_title_text='Count', # yaxis label
        bargap=0.2, # gap between bars of adjacent location coordinates
        bargroupgap=0.1, # gap between bars of the same location coordinates
        width=1000,
        height=600
    )

    fig.show()
```

```python
plotly_compare(ty='hashtags', amt=5)
```

```python
plotly_compare(ty='hashtags', l=['tpmp1000', 'kohlanta'])
```

```python
plotly_compare(ty='hashtags', l=['paris','france'])
```

```python
plotly_compare(ty='emoji', amt=3)
```
